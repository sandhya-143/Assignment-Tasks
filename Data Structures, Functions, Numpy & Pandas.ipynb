### Section A: Data Structures & Control Structures

1. Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# Create a list
numbers = [10, 60, 55, 40, 30]

print("Original List:", numbers)

# append()
numbers.append(35)
print("List of numbers after appending: ", numbers)

#extend()
numbers.extend([70, 90])
print("List of numbers after extend: ",numbers)

#insert()
numbers.insert(5, 79)
print("list after insert: ", numbers)

#remove()
numbers.remove(90)
print("List of numbers after remove: ", numbers)

#pop()
numbers.pop()
print("After pop: ", numbers)

#index()
print("Index of 30: ",numbers.index(30))

#count()
print("Count of 60:", numbers.count(60))

#sort()
numbers.sort()
print("Sorted List:", numbers)

#reverse()
numbers.reverse()
print("Reversed list: ", numbers)

#clear()
temp = numbers.copy()
temp.clear()
print("After clear: ", temp)
```

### Section C: Exploring Pandas

11. Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then: - Use .info(), .describe() - Select 'Name' and 'City' columns - Drop 'City' column - Fill any missing values in 'Salary' column with the mean - Remove any duplicate rows

```python
# Create DataFrame
data = {
    "Name" : ["Anita", "Rahul", "Priya", "Kiran", "Siya", "Sneha", "Anita","Ravi", "Vijay", "Rohan"],
    "Age" : [22, 25, 21, 24, 23, 26, 28, 22, 27, 25],
    "City" : ["Belagavi", "Bengaluru", "Hubli", "Mysuru", "Dharwad", "Belagavi", "Pune", "Hubli","Mysuru", "Bengaluru"],
    "Salary" : [25000, 35000, np.nan, 42000, 30000, 45000, 50000, np.nan, 38000, 35000]
}

df = pd.DataFrame(data)

# Add duplicate row
df.loc[10] = df.loc[1]

print("Original DataFrame: ", df)

print("\n DataFrame Info: ")
print(df.info())

print("\n DataFrame Description: ")
print(df.describe())

print("\n Select Name and City: ")
print(df[["Name", "City"]])

# Drop City Column
df_drop = df.drop(columns=["City"])

print("\n After Dropping City: ")
print(df_drop)

df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

print("\n After Filling Missing Salary: ")
print(df)

# Remove duplicate rows
df = df.drop_duplicates()

print("\n After Removing Duplicate Rows: ")
print(df)
```
