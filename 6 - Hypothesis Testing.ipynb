#### Bombay hospitality Ltd. operates a franchise model for producing exotic Norwegian dinners throughout New England. The operating cost for a franchise in a week (W) is given by the equation W = $1,000 + $5X, where X represents the number of units produced in a week. Recent feedback from restaurant owners suggests that this cost model may no longer be accurate, as their observed weekly operating costs are higher.

```python
import scipy.stats as stats
import numpy as np
import math
from scipy.stats import norm

sample_mean = 3050
X = 600
sample_size = 25
alpha = 0.05
```

```python
population_mean = 1000 + (5 * X)

population_std = 5 * 25
```

#### Step 1: State the Hypotheses

Null Hypothesis (H₀):
The mean weekly operating cost is equal to the theoretical cost (μ = 4000).

Alternative Hypothesis (H₁):
The mean weekly operating cost is greater than the theoretical cost (μ > 4000).

```python
print("Null Hypothesis (H0): µ = 4000 ")
print("Alternative Hypotheis (Ha): µ > 4000")
print()
```

#### Step 2: Calculate the Test Statistic

```python
# Test Statistic
z = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))

print("-----Test Statistic----- \n")
print(f"Sample Mean : {sample_mean}")
print(f"Population Mean : {population_mean}")
print(f"Population standard deviation : {population_std}")
print(f"Sample Size (n): {sample_size}")
print(f"Calculated Z-Statistic: {z:.2f}")
print()
```

#### Step 3: Determine the Probability (p-value)

```python
# Critical Value & P-value

critical_value = norm.ppf(1-alpha)
p_value = 1 - norm.cdf(z)
```

```python
print("-----Probability Comparison----- \n")

print(f"Significance Level: {alpha}")
print(f"Critical Value: {critical_value:.3f}")
print(f"P-value : {p_value: .4f}")
print()
```

#### Step 4: Decision

```python
print("-----Decision-----\n ")

if z > critical_value:
    print("Reject the Null Hypothesis \n ")
else:
    print("Fail to reject the Null Hypothesis \n")
```

```python
print("-----Conclusion----- \n")

if z > critical_value:
    print("There is sufficient evidence to support the restaurant owner's claim")
    print("That the weekly operating costs are higher than the model suggests")

else:
    print("There is insufficient evidence to support the restaurant owner's claim")
    print("That the weekly operating costs are higher than the model suggests")
```

#### Conclusion

Since the p-value is less than the significance level (α = 0.05), the null hypothesis is rejected. Therefore, there is sufficient statistical evidence to conclude that the average weekly operating cost is higher than the theoretical cost predicted by the model. This supports the restaurant owners' claim.

### CONFIDENCE INTERVAL

```python
import numpy as np
from scipy import stats
```

```python
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)
```

```python
print("Sample Size :", n)
print("Sample Mean :", mean)
print("Sample Standard Deviation :", std)
```

#### 99% Confidence Interval Using Sample Standard Deviation (t-distribution)

```python
confidence = 0.99
alpha = 1 - confidence
t_value = stats.t.ppf(1 - alpha/2, df = n - 1)
margin_error = t_value * (std/np.sqrt(n))
lower = mean - margin_error
upper = mean + margin_error

print("99% Confidence Interval using Sample Standard Deviation: ")
print((lower, upper))
```

The t-distribution is used because the population standard deviation is unknown, and the sample size is small (n = 15). When the population standard deviation is not available, the sample standard deviation is used as an estimate, which introduces additional uncertainty. The t-distribution accounts for this uncertainty by having wider tails than the normal distribution, resulting in a more reliable confidence interval for small sample sizes. Therefore, the t-distribution is the appropriate choice for constructing the 99% confidence interval in this problem.

#### Build 99% Confidence Interval Using Known Population Standard Deviation

```python
import numpy as np
from scipy import stats

data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85,
        1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
n = len(data)
mean = np.mean(data)
print("Sample Size: ", n)
print("Sample Mean :", mean)
```

```python
sigma = 0.2
confidence = 0.99
alpha = 1 - confidence
z = stats.norm.ppf(1 - alpha/2)
margin_error = z * (sigma/np.sqrt(n))
lower = mean - margin_error
upper = mean + margin_error

print("99% Confidence Interval :")
print((lower, upper))
```
